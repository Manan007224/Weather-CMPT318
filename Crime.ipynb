```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from haversine import haversine as h
import seaborn as sns
%matplotlib inline
```

```python
filename = 'crime_csv_all_years.csv'
data = pd.read_csv(filename)

years = ['2016','2017']

def year_specific(val) : 
    val = str(val)
    if val in years : 
        return val
    else : 
        return np.nan

to_string = lambda x:str(x)

def get_timestamp(row) : 
    year = row[1]
    month = row[2]
    day = row[3]
    
## only data from 2016 and 2015
data['YEAR'] = data['YEAR'].apply(to_string)
data['YEAR'] = data['YEAR'].apply(year_specific)
data = data.dropna(axis=0,how='any')
```

```python
fig,ax = plt.subplots(figsize=(50,15))
sns.countplot(ax=ax,data=data,x=data['TYPE'])    
plt.savefig('plots/crime_type.jpg')
```
